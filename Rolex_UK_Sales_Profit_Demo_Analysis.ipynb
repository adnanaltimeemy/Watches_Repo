```python
# Title: Rolex UK Sales, Profit Margins & Demographics Dataset
# ---------------------------------------------------------
# This notebook constructs a synthetic dataset of Rolex UK sales,
# profit margins (estimated), and buyer demographics.
# It then exports the dataset to CSV.

## 1) Imports and Setup

import numpy as np
import pandas as pd

# Ensure reproducibility
rng = np.random.default_rng(2026)

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 140)


## 2) Define Basic Data Structure

# Define number of entries to simulate
n_records = 1500

# Create a base DataFrame
df = pd.DataFrame({
    "TransactionID": np.arange(1, n_records+1),
    
    # Simulated billing months over two years
    "Year": rng.choice([2023, 2024], size=n_records, p=[0.45, 0.55]),
    "Month": rng.integers(1, 13, size=n_records),
})


## 3) Simulate Sales Values (£)

# We simulate watch sale prices in GBP — Rolex models generally high-value.
# Use a log-normal distribution to reflect luxury price spread.
df["SalePrice_GBP"] = (rng.lognormal(mean=8.8, sigma=0.4, size=n_records)
                        .round(2))  # e.g., ~£6k–£20k typical range


## 4) Profit Margin and Estimated Profit

# Estimate profit margin % as a synthetic range around 30–40%
df["ProfitMarginPct"] = rng.normal(loc=36, scale=5, size=n_records).clip(15, 60).round(1)

# Calculate profit amounts
df["Profit_GBP"] = (df["SalePrice_GBP"] * (df["ProfitMarginPct"] / 100)).round(2)


## 5) Simulate Buyer Demographics

# Age groups (typical Rolex buyer skew older and affluent)
age_categories = ["Under 30", "30–44", "45–59", "60+"]
age_weights   = [0.08,          0.22,      0.40,      0.30]  # UK luxury skew older

df["BuyerAgeGroup"] = rng.choice(age_categories, size=n_records, p=age_weights)

# Gender distribution — roughly skewed male for luxury watch purchases
df["BuyerGender"] = rng.choice(["Male", "Female", "Non-binary/Other"], 
                               size=n_records,
                               p=[0.70, 0.28, 0.02])

# Region categories in UK
uk_regions = ["London", "South East", "Midlands", "North England", "Scotland", "Wales", "Northern Ireland"]
region_weights = [0.29,       0.20,        0.13,         0.18,              0.10,      0.07,     0.03]
df["BuyerRegion"] = rng.choice(uk_regions, size=n_records, p=region_weights)


## 6) Simulate Purchase Channel

# Assume most purchases are from authorized retailers vs. pre-owned
df["PurchaseChannel"] = rng.choice(["Authorized Dealer", "Pre-Owned Dealer", "Online Boutique"],
                                   p=[0.73, 0.19, 0.08],
                                   size=n_records)


## 7) Optional: Simulate Resale Intent and Trade-In

# Survey data suggests resale/trade patterns among Rolex owners
df["ResaleIntent"] = rng.choice(["Hold Long-Term", "Trade-In", "Sell Within 3y"],
                                p=[0.40, 0.30, 0.30],
                                size=n_records)


## 8) Inspect the Synthetic Dataset

df.head(12)


## 9) Data Summaries (Optional)

print("Summary statistics for Sale Price:")
print(df["SalePrice_GBP"].describe().round(2))

print("\nProfit Margin Distribution (%):")
print(df["ProfitMarginPct"].value_counts(bins=6).sort_index())

print("\nBuyer Demographics by AgeGroup:")
print(df["BuyerAgeGroup"].value_counts(normalize=True).mul(100).round(1))


## 10) Export to CSV

output_file = "rolex_uk_sales_profit_demographics.csv"
df.to_csv(output_file, index=False)

print(f"Export complete: {output_file}")
```
